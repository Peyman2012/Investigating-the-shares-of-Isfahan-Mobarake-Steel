```python
new_df=pd.DataFrame({'Hight':df_folad_day50['<HIGH>'],'Low':df_folad_day50['<LOW>'],
                    'Close':df_folad_day50['<CLOSE>'],'Volume':df_folad_day50['<VALUE>'],'Open':df_folad_day50['<OPEN>']})
```

```python
new_df
```

```python
ndf_folad=np.flip(np.array(new_df))
```

```python
new_df['Diff']=new_df['Open'] - new_df['Close']
```

```python
a=[]
for j in range(50):
    j=0
    a.append(j)
new_df['Buy']=a
new_df['Sell']=a
```

```python
folad_buy=[]
for i in range(len(new_df)):
    if (new_df['Diff'].iloc[i]>0)==True:
        if new_df['Diff'].iloc[i]>0:
            folad_buy.append(new_df['Diff'].iloc[i])
    elif (new_df['Diff'].iloc[i]>0)==False:
        v=0
        folad_buy.append(v)

```

```python
folad_sell=[]
for i in range(len(new_df)):
    if (new_df['Diff'].iloc[i]>=0)==False:
        if new_df['Diff'].iloc[i]<0:
            folad_sell.append(new_df['Diff'].iloc[i])
    elif (new_df['Diff'].iloc[i]>=0)==True:
        v=0
        folad_sell.append(v)
```

```python
new_df['Buy']=folad_buy
new_df['Sell']=folad_sell
```

Investigating the profit or loss of shares

```python
new_df.describe()
```

```python
f=[]
for i in range(len(new_df['Diff'])):
    if (new_df['Diff'].iloc[i])>0:
        f.append('Buy')
    elif (new_df['Diff'].iloc[i])==0:
        f.append('Ziro')
    elif (new_df['Diff'].iloc[i])<0:
        f.append('Sell')
new_df['Signal']=f        
```

```python
new_df
```

```python
plt.bar(range(len(new_df['Buy'])),new_df['Buy'])
plt.bar(range(len(-new_df['Sell'])),-new_df['Sell'])
```

```python
df_per=new_df['Close']//new_df['Close'].iloc[0]
date=df_per.index[1].strftime('%Y-%m-%d')
for i in range(len(df_per)):
    date=df_per.index[i].strftime('%Y-%m-%d')
    if df_per.iloc[i]>=1:
        print(f'The dividend in {date} days is equal to {df_per.iloc[i]}')
    elif df_per.iloc[i]<1:
        print(f'Unfortunately, the dividend in {date} days is equal to {df_per.iloc[i]}')
        
    
```

```python
descending=new_df['Close'][1:]-new_df['Close'].iloc[0]
```

```python
for i in range(len(descending)):
    date=descending.index[i].strftime('%Y-%m-%d')
    if descending.iloc[i]>0:
        print(f'Loss or profit on {date} day is equal to {descending.iloc[i]} which is a loss ')
    elif descending.iloc[i]==0:
        print(f'Loss or profit on {date} day is equal to {descending.iloc[i]} which means that he did not make a profit or a loss')
    elif descending.iloc[i]<0:
        print(f'Loss or profit on {date} day is equal to {descending.iloc[i]} which is  profit')
    
```
